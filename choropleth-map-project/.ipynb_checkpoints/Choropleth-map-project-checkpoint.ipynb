```python
import os
import folium
import json
import pandas as pd
from branca.colormap import linear
import numpy as np
from shapely.geometry import Polygon
from shapely.geometry import Point
from numpy import random
```

```python
# dataset name
dataset_pop_2017 = os.path.join('data', 'population_2017.csv')

# read the data to a dataframe
data2017 = pd.read_csv(dataset_pop_2017)

# eliminate spaces in name of columns
data2017.columns = [cols.replace(' ', '_') for cols in data2017.columns]

data2017.head(10)
```

```python
#lista dos estados do nordeste, siglas
northeast_states = ['AL','BA','CE','MA','PB','PE','PI','RN','SE']

data_states = {}

for state in northeast_states:
    data_states[state] = data2017[data2017['UF'] == state][['NOME_DO_MUNICÍPIO','POPULAÇÃO_ESTIMADA']]    
```

```python
#populacao por estado

pop_per_state = {};

for state in data_states:
    state_cities = data_states[state]
    pop_per_state[state] = state_cities['POPULAÇÃO_ESTIMADA'].sum()

pop_per_state = pd.Series(pop_per_state)
print(pop_per_state)
```

```python
state_file_map = {
    'AL':'geojs-27-mun.json',
    'BA':'geojs-29-mun.json',
    'CE':'geojs-23-mun.json',
    'MA':'geojs-21-mun.json',
    'PB':'geojs-25-mun.json',
    'PE':'geojs-26-mun.json',
    'PI':'geojs-22-mun.json',
    'RN':'geojs-24-mun.json',
    'SE':'geojs-28-mun.json'
}

for state in state_file_map:
    file_founded = os.path.join('geojson', state_file_map[state])
    geo_json_local_data = json.load(open(file_founded,encoding='latin-1'))
    state_file_map[state] = geo_json_local_data  
```

```python
# colormap yellow and green (YlGn)
colormap = linear.YlGn.scale(pop_per_state.min(), pop_per_state.max())
colormap
```

```python
br_states = os.path.join('geojson', 'geojs-states.json')
geo_json_data = json.load(open(br_states,encoding='latin-1'))

# Create a map object
m = folium.Map(
    location=[-5.826592, -35.212558],
    zoom_start=7,
    tiles='Stamen Terrain'
)

# Configure geojson layer
folium.GeoJson(geo_json_data,
              name='Brazil northeast states',
              style_function=lambda feature: {
                  'fillColor': colormap(pop_per_state[feature['properties']['Name']]),
                  'color': 'black',
                  'weight': 1,
                  'dashArray': '5, 5',
                  'fillOpacity': 0.9,
    }).add_to(m)

m
```
